```python
import pandas as pd
import numpy as np
import pickle

impression_df = pd.read_csv('./impressions-train.csv')
final_rating_df = pd.read_csv('./ratings-final.csv')

```

```python
data1 = dict()
data2 = dict()
```

```python
for _, row in impression_df.iterrows():
    reviewerid = row['reviewerid']
    movieid = row['movie-code']
    rating = row['rating']
    if reviewerid not in data1.keys():
        data1[reviewerid] = {movieid: rating}
    else:
        data1[reviewerid][movieid] = rating
        
for _, row in final_rating_df.iterrows():
    reviewerid = row['reviewerid']
    movieid = row['movie-code']
    rating = row['rating-final']
    if reviewerid not in data2.keys():
        data2[reviewerid] = {movieid: rating}
    else:
        data2[reviewerid][movieid] = rating
```

```python
from math import *
def Euclidean(user1,user2,data):
    user1_data=data[user1]
    user2_data=data[user2]
    distance = 0
    for key in user1_data.keys():
        if key in user2_data.keys():
            distance += pow(float(user1_data[key])-float(user2_data[key]),2)
 
    return 1/(1+sqrt(distance))
def top10_simliar(userID,data):
    res = []
    for userid in data.keys():
        if not userid == userID:
            simliar = Euclidean(userID,userid,data)
            res.append((userid,simliar))
    res.sort(key=lambda val:val[1])
    return res[:4]
def recommend(user,data):
    top_sim_user = top10_simliar(user, data)[0][0]
    items = data[top_sim_user]
    recommendations = []
    for item in items.keys():
        if item not in data[user].keys():
            recommendations.append((item,items[item]))
    recommendations.sort(key=lambda val:val[1],reverse=True)
    return recommendations[:5]
```

```python
rec_on_final_rating = {}
rec_on_impression = {}

for i in range(573):
    rec_on_impression[i] = recommend(i,data1)
    rec_on_final_rating[i] = recommend(i,data2)
```
